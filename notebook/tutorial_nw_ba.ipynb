```python

from socpd.model_nw_ba import SocPD 
#from socpd.model_nw_cluster import SocPD as socpdHK



# Visualization
import matplotlib
import matplotlib.pyplot as plt
from socpd.nw_animation import generate_animation,generate_stack
#from socpd.visualization import animate
#import IPython

%matplotlib inline
matplotlib.rcdefaults()
matplotlib.rcParams['animation.embed_limit'] = 2**128
```

```python
dir_params = 'parameters'
all_possible_actions=  [ "Veg_prob",
                        "concerns_animal_welfare",
                        "concerns_convinience",
                        "concerns_environement",
                        "concerns_familiarity",
                        "concerns_health",
                        "concerns_hunger",
                        "concerns_price",
                        "homo_om_inf",
                        "homo_veg_inf"

                        ]

Hypothesis_settings = { 'dir_params' : dir_params , # folder name stored actions parameters
                        'status_var':'vegetarian',  # the variables deciding agents' status
                        # all actions from the actions parameters
                        'all_possible_actions':all_possible_actions,
                        # all actions/events influencing directly on the agent
                        'actions_to_self':["Veg_prob",
                                          "concerns_animal_welfare",
                                          "concerns_convinience",
                                          "concerns_environement",
                                          "concerns_familiarity",
                                          "concerns_health",
                                          "concerns_hunger",
                                          "concerns_price",],
                                            
                        # all actions/events from the agent influencing on others in thier networks
                        'actions_to_nw':['homo_veg_inf',
                                        'homo_om_inf',],

                        # actions/events from the agents having negative effects on thier networks 
                        'actions_to_nw_neg':['homo_om_inf'],
                        # actions/events from the agents having positive effects on thier networks 
                        'actions_to_nw_pos':['homo_veg_inf',],
                        }
 


parameters = {'Hypothesis_settings' : Hypothesis_settings,
            'steps': 60,    
            'seed' : 123456,  
            'pop' : 10000, # size of the population
            'm': 1, # Number of edges with which a new node attaches to existing nodes
            'p' : 0, # each existing nodes has p prob to form m new link(s) to the others with attachment preference
            'q' : .7, # each existing nodes has q prob to rewire m existing link(s) with attachment preference
            'use_ipf': True,   
            } 
```

```python
model = SocPD(parameters)
result= model.run()
result.reporters
```

```python
ani_dict = {'stack_plot_name': "Diet transition",
            'pos_neg_label_list': ['vegetarians', 'omnivores'], 
            'model_name': 'SocPD', 
            "out_path":"output/BA_stackplot"}

  
generate_stack(SocPD, parameters,ani_dict)
```

```python
ani_dict = {'graph_nw_name': 'Diet difusion in network',
            'stack_plot_name': "Diet transition",
            'pos_neg_label_list': ['vegetarians', 'omnivores'], 
            'model_name': 'SocPD',
            'width': 0.1,
            'output_name': 'output/tuned_BA_network'}

generate_animation(SocPD, parameters, ani_dict)
```
